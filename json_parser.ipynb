```python
import json
import csv

with open('messages.json', 'r') as dataset:
    message_data = json.load(dataset)

    # Access the "messages" key which contains a list of messages
    messages = message_data.get("messages", [])

    # Dictionary to store the first occurrence of each topic_id
    first_occurrence_dict = []

    # Iterate through each message and append the data to the list
    for message in messages:
        topic_id = message.get("topic_id")
        if topic_id not in [entry["topic_id"] for entry in first_occurrence_dict]:
            first_occurrence_dict.append({
                "topic_id": topic_id,
                "message": message.get("text")
            })

    # Write the data to a CSV file
    csv_file_path = 'output.csv'
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        fieldnames = ["topic_id", "message"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        # Write header
        writer.writeheader()

        # Write data
        for entry in first_occurrence_dict:
            writer.writerow(entry)

#print(f"Output has been saved to {google_chat.csv}")
```

```python
import json
import csv

with open('messages.json', 'r') as dataset:
    message_data = json.load(dataset)

    # Access the "messages" key which contains a list of messages
    messages = message_data.get("messages", [])

    # Dictionary to store the first occurrence of each topic_id
    first_occurrence_dict = []

    # Iterate through each message and append the data to the list
    for message in messages:
        topic_id = message.get("topic_id")
        if topic_id not in [entry["topic_id"] for entry in first_occurrence_dict]:
            creator = message.get("creator", {})
            name = creator.get("name", "")
            created_date = message.get("created_date", "")
            
            first_occurrence_dict.append({
                "topic_id": topic_id,
                "name": name,
                "created_date": created_date,
                "message": message.get("text")
            })

    # Write the data to a CSV file
    csv_file_path = 'output.csv'
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        fieldnames = ["topic_id", "name", "created_date", "message"]
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        # Write header
        writer.writeheader()

        # Write data
        for entry in first_occurrence_dict:
            writer.writerow(entry)

print(f"Output has been saved to {csv_file_path}")
```
